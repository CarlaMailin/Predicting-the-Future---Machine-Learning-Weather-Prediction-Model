```python
# Import packages
import xarray as xr
import copy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

import optuna
optuna.logging.set_verbosity(optuna.logging.WARNING)

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

import xarray as xr

# Open dataset
# ds = xr.open_dataset("thess_sept.nc")
ds = xr.open_dataset("thess_sept.nc", engine="netcdf4")
print(ds)

# Extract variables
temp_c = ds["t2m"] - 273.15   # Kelvin to Celsius
lat = ds["latitude"]
lon = ds["longitude"]
time = ds["valid_time"]

# Quick check
print(temp_c.shape)
print(time.values[0], time.values[-1])

# Select nearest point to Thessaloniki
thess_lat = 40.64
thess_lon = 22.94

temp_thess = temp_c.sel(
    latitude=thess_lat,
    longitude=thess_lon,
    method="nearest"
)

print(temp_thess.shape)

# Convert to 1D numpy array
temp_1d = temp_thess.values
print(temp_1d)
```
